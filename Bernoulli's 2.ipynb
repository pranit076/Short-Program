A siphon of diameter 200 mm connects two reservoirs having a difference of elevation of 20
m. The total length of siphon is 500m and the summit is 3m above the water level in the
upper reservoir. The length of siphon from upper reservoir to summit is 100 m. Take
friction factor = 0.0005. Determine the speed. Note: Neglect minor loss

```python
from math import pow, sqrt, pi
import numpy as np
```

Calculating the velocity, given formula v = (z*g*d/2*F*l1)^1/2

```python
d = float(input("Enter the diameter of syphon in meter : "))
za = float(input("Enter Datum difference from point a to b in meter : "))
zb = float(input("Enter the elevation of point below resoviour in meter : "))
la = float(input("Enter the length of the siphon in meter : "))
F = 0.005 #coefficiet of friction
g = 9.8 #accleration due to gravity
```

```python
z = (za - zb)
print("The height difference is: ", z)
def velo(a,b,c,h): 
    s = (z*g*d)/(2*F*la)
    v = sqrt(s)
    return v
```

```python
velocity_at_summit = velo(g, z, d, F)
print(velocity_at_summit)
```

```python
r = d/2
A = pi*pow(r,2)
def flow_rate(x,y): 
    Q = A * velocity_at_summit
    return Q
```

```python
flowrate = flow_rate(A, velocity_at_summit)
print("The flow rate is : ", flowrate)
```

```python
PHA = 10.33 #pressure_head_at_point_a
VHA = 0 #velocity_at_point_a
lb = float(input("Enter length of siphon from upper reservoir to summit, in meter:"))
zs = 3.0 #datum_difference_from_point_S_to_A
```

```python
vs = float(pow(velocity_at_summit, 2))
print("velocity at summit:", vs)
hf = float(4*F*lb*vs)/(2*g*d) #major_loos_or_energy_loss
print("The hf is:", hf)
def pressure_head_at_s(x,y,z,w,r,t,i,u,o): 
    PHS = PHA + VHA - zs - vs/(2*g) - hf #za is datum head of point a, thus taken 0. 
    return PHS
```

```python
pressure_head_at_summit = pressure_head_at_s(PHA, za, VHA, zs, vs, g, F, lb, d)
print(pressure_head_at_summit)
```
